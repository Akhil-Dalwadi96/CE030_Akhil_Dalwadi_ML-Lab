```python
#ex-9

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('Iris.csv')

arr1= np.array(data['Id'])
nparrayCentered = arr1 - np.mean(arr1)
print("For ID:\n", nparrayCentered)

arr1= np.array(data['SepalLengthCm'])
nparrayCentered = arr1 - np.mean(arr1)
print("For SepalLengthCm:\n", nparrayCentered)

arr1= np.array(data['SepalWidthCm'])
nparrayCentered = arr1 - np.mean(arr1)
print("For SepalWidthCm:\n",nparrayCentered)

arr1= np.array(data['PetalLengthCm'])
nparrayCentered = arr1 - np.mean(arr1)
print("For PetalLengthCm:\n",nparrayCentered)

arr1= np.array(data['PetalWidthCm'])
nparrayCentered = arr1 - np.mean(arr1)
print("For PetalWidthCm:\n",nparrayCentered)
```

```python
#ex-10

import numpy as np
matrix = np.array([[1,1,1],[2,2,2],[3,3,3]])
matrix1 = np.array([[1,2,3],[3,4,5],[4,3,1]])
print('Original matrix:\n',matrix)

print('Transposed matrix:\n',matrix)

print('Norm:\n',np.linalg.norm(matrix))

print('Multiplication of two matrix:\n',np.dot(matrix,matrix1))

print('Sum:\n',np.sum(matrix * matrix1))

print('Matrix mean:\n',np.mean(matrix))
print('Mean by columns:\n',np.mean(matrix,axis=0))
print('Mean by rows:\n',np.mean(matrix,axis=1))

matrixCentered = matrix-np.mean(matrix, axis=0)
print('Centered by columns matrix:\n',matrixCentered)
print('New mean by column:\n',matrixCentered.mean(axis=0))
```
